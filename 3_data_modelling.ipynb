## Introduction

### This script:
- Builds the land development model scenarios for the study area using raster layers contained in the "inputs_TUR" folder.
- Normalizes or inverts values in the raster layers as required.
- Calculates a regression equation combining the values of every raster layer for each 10 metre by 10 metre pixel.
- Repeats the regresion equation for each model scenario below.


### Model Equations:

##### Land Suitability:
$$
LS = (\frac{Negative Attractor}{Total Attractors}\cdot{Normalized Negative Attractor})\cdot
({Constraints}\cdot{Constraints})
$$
 
##### Land Desireability:
$$
LD = (\frac{Positive Attractor}{Total Attractors}\cdot{Normalized Positive Attractor}))\cdot 
({Constraints}\cdot{Constraints})
$$

##### Land Suitability and Desireability:
$$
LSD = ((Land Suitability) + (Land Desireability)) \cdot ({Constraints}\cdot{Constraints})
$$

## PART 1 - Import Raster Layers

All raster layers are classified according to six categories:
    
| **Category**             	| **Number of Layers** 	| **Purpose**                                                                                                                                                                      	| **Equation Variables**             	| **Source**       	|
|:-------------------------	|:--------------------:	|:---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------	|:--------------------------        	|:-----------------	|
| study boundary        	| 1                    	| Covers the study area; assigns a value of 0 to all pixels outside the layer; Pixels within the layer remain developable                                                          	| constant for all scenarios        	| LUTI             	|
| structural constraints  	| 3                   	| Covers the area where development is not possible; Assigns a value of 0 to all pixels within the layer; Pixels outside the layer remain developable                              	| constant for all scenarios          	| BDTRE         	|
| conservation constraints 	| 5                   	| Covers the area where development is not possible; Assigns a value of 0 to all pixels within the layer; Pixels outside the layer remain developable                              	| scenarios 4 to 7                  	| PAI; PPR; PTC2  	|
| land use constraints  	| 5                   	| Covers the area where development is not possible; Assigns a value of 0 to all pixels within the layer; Pixels outside the layer remain developable                              	| scenarios 6 to 7                  	| PRG              	|
| negative attractors   	| 1                    	| Classifies development suitability; Assigns a value on a scale of 0 to 1 to all pixels in the layer; Higher values are less suitable; Values require normalization and inversion 	| constant for all scenarios        	| DEM              	|
| positive attractors   	| 12                   	| Classifies development desirability; Assigns a value on a scale of 0 to 1 to all pixels in the layer; Higher values are more desirable; Values require normalization             	| varies by year for scenarios 2 to 7	| LUTI             	|

```python
# import raster data with rxr.open_rasterio()

# boundary layers
bound = rxr.open_rasterio(inputs["TurinFUA"])

# structural constraint layers
road = rxr.open_rasterio(inputs["BDTREroadcover"])
rail = rxr.open_rasterio(inputs["BDTRErailcover"])
water = rxr.open_rasterio(inputs["BDTREwatercover"])

# conservation constraint layers
unesco = rxr.open_rasterio(inputs["PTCunesco"])
urbangreen = rxr.open_rasterio(inputs["PTCurbangreen"])
forestcover = rxr.open_rasterio(inputs["PTCforestcover"])
conservation = rxr.open_rasterio(inputs["PPRconservationarea"])
floodzone = rxr.open_rasterio(inputs["PAIfloodzone"])

# land use constraint layers
parksgardens = rxr.open_rasterio(inputs["PRGparksandgardens"])
industry = rxr.open_rasterio(inputs["PRGindustry"])
health = rxr.open_rasterio(inputs["PRGhealthcare"])
education = rxr.open_rasterio(inputs["PRGeducation"])
cemetery = rxr.open_rasterio(inputs["PRGcemetery"])

# negative attractor layers
slope = rxr.open_rasterio(inputs["DEMslope"])

# positive attractor layers
job19c = rxr.open_rasterio(inputs["JobAccessCar19"])
job19b = rxr.open_rasterio(inputs["JobAccessBus19"])
job19r = rxr.open_rasterio(inputs["JobAccessRail19"])
job30c = rxr.open_rasterio(inputs["JobAccessCar30"])
job30b = rxr.open_rasterio(inputs["JobAccessBus30"])
job30r = rxr.open_rasterio(inputs["JobAccessRail30"])
house19c = rxr.open_rasterio(inputs["HousingAccessCar19"])
house19b = rxr.open_rasterio(inputs["HousingAccessBus19"])
house19r = rxr.open_rasterio(inputs["HousingAccessRail19"])
house30c = rxr.open_rasterio(inputs["HousingAccessCar30"])
house30b = rxr.open_rasterio(inputs["HousingAccessBus30"])
house30r = rxr.open_rasterio(inputs["HousingAccessRail30"])

print("Complete")
```

## PART 2 - Normalize Data

```python
# normalise negative and positive attractors

# slope
slope_n = slope / 78.8661  # divide by the highest slope degree within the study area

# job accessibility
job19c_n = job19c / np.nanmax(job19c.values)
job19b_n = job19b / np.nanmax(job19b.values)
job19r_n = job19r / np.nanmax(job19r.values)
job30c_n = job30c / np.nanmax(job30c.values)
job30b_n = job30b / np.nanmax(job30b.values)
job30r_n = job30r / np.nanmax(job30r.values)

# housing accessibility
house19c_n = house19c / np.nanmax(house19c.values)
house19b_n = house19b / np.nanmax(house19b.values)
house19r_n = house19r / np.nanmax(house19r.values)
house30c_n = house30c / np.nanmax(house30c.values)
house30b_n = house30b / np.nanmax(house30b.values)
house30r_n = house30r / np.nanmax(house30r.values)

print("Complete")
```

## PART 3 - Model Scenario 1

| **Model** | **Scenario**  | **Description**                                            | **Attractors** | **Constraints** | **LS** | **LSD** |
|:---------:|:--------------|:-----------------------------------------------------------|:--------------:|:---------------:|:------:|:-------:|
|     1     | csLS          | City Structure                                             |        1       |        4        |    x   |         |


#### Land Suitability

```python
# set attractors
n_of_attractors = 1

# assign weight
slope_w = 1 / n_of_attractors

# equation
csLS = ((slope_w * (1 - slope_n)) * (
    bound * rail * road * water))
csLS <- csLS.compute()
```

```python
f = plt.figure()
f.set_size_inches(8, 6)
#f.suptitle(f"Scenario", ha='center', size=20)

ax1 = plt.subplot(1, 1, 1) # columns, rows, location
csLS.plot(ax=ax1, cmap="Blues", vmin=0, vmax=1)
tur_luti.boundary.plot(ax=ax1, linewidth=0.5, edgecolor='grey')

scalebar = ScaleBar(0.08, "m", length_fraction=0.25, 
                    location='lower right', 
                    scale_loc='top',
                    box_color='none')
ax1.add_artist(scalebar)

plt.title("Model 1 - csLS", fontsize=14, color='black', loc='center', pad=16)
plt.xticks([])
plt.yticks([])
plt.xlabel("")
ax1.yaxis.set_label_position("right")
plt.ylabel("Development Potential", fontsize=14, labelpad=10)

f.tight_layout()

plt.savefig(outputs["Scenario1"], dpi=2000)

print("Done.")
```

## PART 4 - Model Scenario 2

|     2     | csLSD19       | City Structure & 2019 LUTI                                 |        7       |        4        |        |    x    |


#### Land Suitability and Desirability

```python
# set attractors
n_of_attractors = 7

# assign weight
slope_w = 1 / n_of_attractors
job19b_w = 1 / n_of_attractors
job19r_w = 1 / n_of_attractors
job19c_w  = 1 / n_of_attractors
house19b_w = 1 / n_of_attractors
house19r_w = 1 / n_of_attractors
house19c_w = 1 / n_of_attractors

# equation
csLSD19 = ((slope_w * (1 - slope_n)) + (
    job19b_w * job19b_n) + (job19r_w * job19r_n) + (job19c_w * job19c_n) + (house19b_w * house19b_n) + (house19r_w * house19r_n) + (house19c_w * house19c_n)) * (
    bound * rail * road * water)
csLSD19.compute()
```

## Clip Rasters to Selected Municipalities

```python
# check crs
print("tur_loc", tur_loc.crs)
print("zones_loc", zones_loc.crs)
```

```python
# SOURCE FOR CODE: GeoDev YoutTube channel - https://www.youtube.com/watch?v=TbbL2yeRTsk

# open desired boundary with fiona to be able to clip
# we will use the municipal study area
with fiona.open('clean_data/tur_loc.gpkg', 'r') as layer:
    for feature in layer:
       shapes = [feature['geometry']]
```

```python
# open with rasterio to actually do the clipping

# first the 2019 results
with rasterio.open('outputs_TUR/csLSD19_wPALU.tif') as src:
    out_image, out_transform = rasterio.mask.mask(src, shapes, crop=True)
    out_meta = src.meta

out_meta.update({
    "driver": 'Gtiff',
    'height': out_image.shape[1],
    'width': out_image.shape[2],
    'transform': out_transform
})

with rasterio.open('outputs_TUR/muni_csLSD19_wPALU.tif', 'w', **out_meta) as dst:
    dst.write(out_image)
```

```python
# check to confirm
show(rasterio.open('outputs_TUR/muni_csLSD19_wPALU.tif'))
```

```python
# then the 2030 results
with rasterio.open('outputs_TUR/csLSD30_wPALU.tif') as src:
    out_image, out_transform = rasterio.mask.mask(src, shapes, crop=True)
    out_meta = src.meta

out_meta.update({
    "driver": 'Gtiff',
    'height': out_image.shape[1],
    'width': out_image.shape[2],
    'transform': out_transform
})

with rasterio.open('outputs_TUR/muni_csLSD30_wPALU.tif', 'w', **out_meta) as dst:
    dst.write(out_image)
```

```python
# check to confirm
show(rasterio.open('outputs_TUR/muni_csLSD30_wPALU.tif'))
```

```python
# and finally the increase in development potential
with rasterio.open('outputs_TUR/csLSD_wPALU_1930.tif') as src:
    out_image, out_transform = rasterio.mask.mask(src, shapes, crop=True)
    out_meta = src.meta

out_meta.update({
    "driver": 'Gtiff',
    'height': out_image.shape[1],
    'width': out_image.shape[2],
    'transform': out_transform
})

with rasterio.open('outputs_TUR/muni_csLSD_wPALU_1930.tif', 'w', **out_meta) as dst:
    dst.write(out_image)
```
